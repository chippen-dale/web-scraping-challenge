```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import pandas as pd
```

```python
#Site for Scraping
url = 'https://mars.nasa.gov/news/'
response = requests.get(url)
```

```python
# Parsing with Beautiful Soup
soup = bs(response.text, 'html.parser')
```

```python
#Filter Title div
title_results = soup.find_all('div', class_="content_title")
title_results
```

```python
# Create list for storing
mars_titles = []
# Loop over div elements
for title in title_results:
    # Identify the anchor...
    if (title.a):
        # And the anchor has non-blank text...
        if (title.a.text):
            # Append thext to the list
            mars_titles.append(title)
mars_titles
```

```python
#Strip the headlines and clean list of titles
news_title = []
# Print only the headlines
for x in range(6):
    clean_list=mars_titles[x].text
    final_list = clean_list.strip('\n\n')
    news_title.append(final_list)
news_title
```

```python
#Find paragraph description
paragraph_results = soup.find_all('div', class_="rollover_description_inner")
paragraph_results
```

```python
news_p = []
# Strip and only keep paragraph text. 
for x in range(6):
    clean_p_list=paragraph_results[x].text
    final_p_list = clean_p_list.strip('\n\n')
    news_p.append(final_p_list)
news_p
```
